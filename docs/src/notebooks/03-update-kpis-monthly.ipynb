# Update KPIs Monthly

### List KPIs

```python

import requests

def list_data_rows():
    url = "http://localhost:8080/api/v2/analytics/data.csv"

    payload = ""
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
    }

    response = requests.request("GET", url, headers=headers, data=payload)
    response.raise_for_status()
    return response.json()
```

### Update KPI this month

```python
def add_new_month_of_data():
    import requests
    import json

    url = "http://localhost:8080/api/v2/analytics/data.csv"

    payload = json.dumps({
    "year": 2026,
    "month": 2,
    "user_satisfaction_score": "3.1/5",
    "digital_take_up_percentage": "82%",
    "cost_per_transaction": "£3.11",
    "order_completion_rate_percentage": "40%"
    })
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'Content-Type': 'application/json'
    }

    response = requests.request("POST", url, headers=headers, data=payload)

    print(response.text)
```
